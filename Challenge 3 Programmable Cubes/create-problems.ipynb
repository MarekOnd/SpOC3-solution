```python
import numpy as np
import json
import os
```

```python
PROBLEM_NAME = "test"

initial_config = np.array([[0,0,0],[1,0,0],[-1,0,0]])
target_config  = np.array([[0,0,0],[0,1,0],[-1,0,0]])
initial_types = np.array([0,0,1])
target_types = np.array([0,0,1])
data = {
 "path": f"data/{PROBLEM_NAME}",
 "plot_dim": 5,
 "max_cmds": 100, 
 "num_cubes": 3,
 "num_cube_types": 2,
 "fitness_offset": 0.66,
 "colours": ["lightgray",
            "limegreen"]
}
json_data = json.dumps(data,indent=4)
try:
    os.mkdir(f"data/{PROBLEM_NAME}")
except:
    print("directory already exists")
with open(f"problems/{PROBLEM_NAME}.json","w") as f:
    f.write(json_data)
np.save('data/{}/Target_Config.npy'.format(PROBLEM_NAME),target_config)
np.save('data/{}/Initial_Config.npy'.format(PROBLEM_NAME),initial_config)
np.save('data/{}/Target_Cube_Types.npy'.format(PROBLEM_NAME),target_types)
np.save('data/{}/Initial_Cube_Types.npy'.format(PROBLEM_NAME),initial_types)
```
