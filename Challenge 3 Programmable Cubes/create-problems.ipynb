## Inverse problems

```python
PROBLEM_NAME = "JWST"
udp = programmable_cubes_UDP(PROBLEM_NAME)
udp.fitness(np.array([-1]))
ti = np.array(udp.initial_cube_types)
ci = np.array(udp.final_cube_positions)
ct = np.array(udp.target_cube_positions)
tt = np.array(udp.target_cube_types)
types = np.arange(np.max(ti)+1)

# invert

INVERTED_PROBLEM_NAME = PROBLEM_NAME + "_INV"


json_data = ""
with open(f"problems/{PROBLEM_NAME}.json","r") as f:
    json_data += f.read()
data = json.JSONDecoder().decode(json_data)
data["path"] = f"data/{INVERTED_PROBLEM_NAME}"
json_data = json.dumps(data,indent=4)

try:
    os.mkdir(f"data/{INVERTED_PROBLEM_NAME}")
except:
    print("directory already exists")

with open(f"problems/{INVERTED_PROBLEM_NAME}.json","w") as f:
    f.write(json_data)

np.save('data/{}/Target_Config.npy'     .format(INVERTED_PROBLEM_NAME),ci)
np.save('data/{}/Initial_Config.npy'    .format(INVERTED_PROBLEM_NAME),ct)
np.save('data/{}/Target_Cube_Types.npy' .format(INVERTED_PROBLEM_NAME),ti)
np.save('data/{}/Initial_Cube_Types.npy'.format(INVERTED_PROBLEM_NAME),tt)
```
